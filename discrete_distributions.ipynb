```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import seaborn as sns
sns.set()
```

```python
# plot binomial distribution
def binomial_distribution(x, p, n):
    label = f"p={p} and n={n}"
    coeff = factorial(n) // factorial(x) // factorial(n-x)
    y = coeff * (p ** x) * (1-p) ** (n-x)
    plt.xlim(0, 40)
    plt.ylim(0, 0.25)
    sns.scatterplot(x, y, label=label)

x1 = np.arange(21)
x2 = np.arange(41)

binomial_distribution(x1, 0.5, 20)
binomial_distribution(x1, 0.7, 20)
binomial_distribution(x2, 0.5, 40)

plt.title("Binominal distribution")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```

```python
def geometric_distribution(x, p):
    label = f"p={p}"
    y = ((1 - p) ** (x-1)) * p
    plt.xlim(0, 10)
    plt.ylim(0, 1.0)
    sns.lineplot(x, y, label=label)

x = np.arange(1, 11)
geometric_distribution(x, 0.2)
geometric_distribution(x, 0.5)
geometric_distribution(x, 0.8)

plt.title("Geometric distribution")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```

```python
def poisson_distribution(x, l):
    label=f"$\lambda$={l}"
    y = (l ** x) * np.e ** (-l) / factorial(x)
    plt.xlim(0, 20)
    plt.ylim(0, 0.4)
    sns.lineplot(x, y, label=label)

x = np.arange(0, 21)

poisson_distribution(x, 1)
poisson_distribution(x, 4)
poisson_distribution(x, 10)

plt.title("Poisson distribution")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```
