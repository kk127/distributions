```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import seaborn as sns
sns.set()
```

```python
def gaussian_distribution(x, mu, sigma):
    label=f"$\mu$={mu}, $\sigma$={sigma}"
    coef = 1 / np.sqrt(2 * np.pi * sigma ** 2)
    power = - ((x - mu) ** 2) / (2 * sigma ** 2)
    y = coef * np.e ** power
    sns.lineplot(x, y, label=label)

x = np.linspace(-6, 6, 100)

gaussian_distribution(x, 0, 1)
gaussian_distribution(x, 0, 0.5)
gaussian_distribution(x, 0, 2.5)
gaussian_distribution(x, -2, 0.7)

plt.title("Gaussian distribution")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.plot()
```

```python
def exponential_distribution(x, l):
    label=f"$\lambda$={l}"
    y = l * pow(np.e, - l * x)
    sns.lineplot(x, y, label=label)

x = np.linspace(0, 5, 100)

exponential_distribution(x, 0.5)
exponential_distribution(x, 1.0)
exponential_distribution(x, 1.5)

plt.title("Exponential distribution")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.plot()
```
